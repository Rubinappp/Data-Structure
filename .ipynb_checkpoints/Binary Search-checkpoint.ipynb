Sort and pick the middle and determine if target card lies left or right.eliminate half array and repeat the step.

```python
def locate_card(cards,query):   
    low=0
    high=len(cards)-1
    while low<=high:
        position=(low+high)//2
        mid=cards[position]
        print('low',low,' mid_number',mid,' high',high,' mid',position)
        if(mid==query):
            return position
        elif(mid>query):
            low=position+1
        elif(mid<query):
            high=position-1
    return -1
cards=[2,10,4,7,9]
cards.sort(reverse=True)
cards
```

```python
locate_card(cards,10)
```

Here my algorithm :

1. Uses binary search to find one occurrence.
2. Walk left one by one until the first occurrence is found.

Time complexity =O(logn)+O(n)

not optimized solution

The better approach:

Instead of walking left after finding the query, keep using binary search.

When you find the query:

- Save the position.
- Continue searching the left half to see if there's an earlier occurrence.

```python
def test_location(cards,query,position):
    tl=0   
    th=position
    while(tl<=th):
        mid=(tl+th)//2
        mid_num=cards[mid]
        print('low',tl,' mid_number',mid_num,' high',th,' mid',mid)
        if(mid_num==query):
            if(mid==0 or cards[mid-1]!=query):
                return mid
            th=mid-1
        elif(mid_num!=query):
            tl=mid+1
            
def locate_card(cards,query):   
    low=0
    high=len(cards)-1
    while low<=high:
        position=(low+high)//2
        mid_num=cards[position]
        print('low',low,' mid_number',mid_num,' high',high,' mid',position)
        if(mid_num==query):
            if(position>0 and cards[position-1]==query):
                position=test_location(cards,query,position)
            return position
        elif(mid_num>query):
            low=position+1
        elif(mid_num<query):
            high=position-1
    return -1

cards=[2,2,1,1,8,9,3,6,3,10,10,10, 9, 8]
cards.sort(reverse=True)
cards
```

```python
locate_card(cards,10)
```

This is our standard solution.
O(logn)

lets  make it more easier by seperating function and its tasks

```python
def test_cases(cards,mid,query):
    if(cards[mid]==query):
        if(mid-1>=0 and cards[mid-1]==query):
            return 'left'
        else:
            return 'found'
    elif(cards[mid]>query):
        return 'right'
    elif(cards[mid]<query):
        return 'left'


def locate_cards(cards,query):
    h=len(cards)-1
    l=0
    while(h>=l):
        mid=(h+l)//2

        result=test_cases(cards,mid,query)
        
        if(result=='found'):
            return mid
        elif(result == 'left'):
            h=mid-1
        elif(result=='right'):
            l=mid+1
    return -1
cards=[10,10,2,2,1,1,8,9,3,6,3,10,10,10, 9, 8]
cards.sort(reverse=True)
cards
```

```python
locate_cards(cards,10)
```

We check execution time of both linear and binary search to see the difference.

```python
large_test=({
'input':{
    'cards':list(range(100000,0,-1)),
     'query':1
},
    'output':99999
})
```

```python
list(range(10000,0,-1))
```

```python
def linear_locate_card(cards,query):
    position=0
   
    while position<len(cards):
    
        if(cards[position] == query):
            return position
        position+=1
    return -1
```

```python
result,passed,runtime=evaluate_test_case(linear_locate_card,large_test,display=False)
print(f'result:{result}\npassed:{passed}\nexecution time:{runtime} ms')
```

```python
result,passed,runtime=evaluate_test_case(locate_cards,large_test,display=False)
print(f'result:{result}\npassed:{passed}\nexecution time:{runtime} ms')
```

In worst case my linear search has around 1 million check while my binary search has around log2(100000)=17 checks.

For list 10 times ,linear search runs 10 times longer whereas binary search would require 3 additional operations only. 


if we divide linear time complexity with binary.
we can say binary search is c*N/log2N times faster than linear search
