### Problem:Rotated lists

You are given list of numbers, obtained by rotating a sorted list an unknown
number of times. Write a function to determine the minimum number of times the
original sorted list was rotated to obtain the given list. Your function should have
the worst-case complexity of 0(log N) , where N is the length of the list. You can
assume that all the numbers in the list are unique.

Example: The list [5, 6, 9, 0, 2, 3, 4] was obtained by rotating the
sorted list [0, 2, 3, 4, 5, 6, 9] 3 times.

We define "rotating a list" as removing the last element of the list and adding it
before the first element. E.g. rotating the list [3, 2, 4, 1] produces [1, 3,
2,4].

"Sorted list" refers to a list where the elements are arranged in the increasing order
e.g. [1, 3, 5, 7] .

#### Linear search solution.
#### 3. comming up with solution in plain text

if rotated k times, smallest number of list will be in k position starting from 0


1. create position=0
2. compare numbers after it.
3. if after number is smaller then return position+1
4. otherwise increase position and repeat from 2

(Notice if we worked from last position we need to check if our predecessor is greater than current element)

#### 4. Implementation

```python
def curr_rotations_linear(nums):
    position=0
    while(position<len(nums)):
        if nums[position+1]<nums[position]:
            return position+1
        position+=1
    return 0
curr_rotations_linear([10,20,30,35,1,5,8])
```

```python
def curr_rotations_linear(nums):
    position=1
    while(position<len(nums)):
        if position>0 and nums[position]<nums[position-1]:
            return position
        position+=1
curr_rotations_linear([10,20,30,35,1,5,8])
```

Time complexity=O(n) and space complexity=O(1)

#### Binary Search

#### comming up with solution in plain text

if rotated k times, smallest number of list will be in k position starting from 0

and before numbers are all higher than number in k position.
nums[mid-1]>nums[mid]

1. if middle element is smaller than its predecessor than thats the answer.

2. Question is if i have 
    1. nums=40,30,20,1,2,3 ( ans lies right of middle element)
    2. nums=40,30,1,2,5,9,12 (ans lies left of middle element)

- we can see 20 is middle element and middle element> last element
then search in right side .
- we can see 2 is middle element and middle element< last element then search is left side


if middle element is smaller or equal to  predecessor return mid index
otherwise check if middle element is less than last element if yes.
Then apply binary search on right position
otherwise apply binary search on left position
#### Implementation

```python
def count_rotations(nums):
    l=0
    h=len(nums)-1
    if len(nums)== 0:
        return -1
    while(l<=h):
        mid=(l+h)//2
        if(nums[mid-1]>=nums[mid]):
            return mid  
        if(nums[mid]>nums[-1]):
            l=mid+1
        if(nums[mid]<nums[-1]):          
             h=mid-1
```
