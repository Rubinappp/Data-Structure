Notice i can only find one position at a time.
- For the first occurrence, the rule is:
Is there another query on the left?
- For the last occurrence, the rule is:
Is there another query on the right?

- One algorithm wants to move left.

- The other wants to move right.

A single binary search cannot obey both rules simultaneously.

### Lets write in more generic way

```python
def binary_search(l,h,condition):
    while(l<=h):
        mid=(l+h)//2
        result=condition(mid)
        if(result=='found'):
            return mid
        elif(result == 'right'):
            l=mid+1
        else:
            h=mid-1
```

```python
def first_position(query,cards):
    def condition(mid):
        if(cards[mid]==query):
            if(mid>=0 and cards[mid-1]==query):
                return 'left'
            else:
                return 'found'
        elif(cards[mid]>query):
            return 'right'
        else:
            return 'left'
    return binary_search(0,len(cards)-1,condition)
```

```python
def last_position(query,cards):
    def condition(mid):
        if(cards[mid]==query):
            if(mid<len(cards)-1 and cards[mid+1]==query):
                return 'right'
            else:
                return 'found'
        elif(cards[mid]>query):
            return 'right'
        else:
            return 'left'
    return binary_search(0,len(cards)-1,condition)
```

```python
def first_and_last_position(query,cards):
    return first_position(query,cards),last_position(query,cards)
```
