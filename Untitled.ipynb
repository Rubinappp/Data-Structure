Why  first occurrence?

Because if duplicates exist and the answer can be any index, there isn't a unique expected output. output is non deterministic( multiple indexes).
Asking for the first occurrence makes the result deterministic, which simplifies testing and ensures every correct implementation produces the same answer.

3. correct solution of problem in plain language.

we can use brute force soultion checking all possible answer.

1. create a variable position=0.
2. check position in card equals to query
3. if it is equal then we get the ans so return it
4. if not equal increase position by 1 and repeat 2 and 3
5. if number not found till the end return -1

linear search algorithm.

Implementation and testing using test cases.

```python
def locate_card(cards, query):
    position = 0
    for i in range(0, len(cards)):
        if(len(cards) == 0):
            return -1
        if(cards[position] == query):
            return position
        if(position == (len(cards)-1)):
            return -1
        if(cards[position] != query):
            position += 1
```

```python
def locate_card(cards, query):
    position = 0
    while True:
        if(cards[position] == query):
            return position     
        position += 1
        if(position == len(cards)):
            return -1
```
