Sort and pick the middle and determine if target card lies left or right.eliminate half array and repeat the step.

```python
def locate_card(cards,query):   
    low=0
    high=len(cards)-1
    while low<=high:
        position=(low+high)//2
        mid=cards[position]
        print('low',low,' mid_number',mid,' high',high,' mid',position)
        if(mid==query):
            return position
        elif(mid>query):
            low=position+1
        elif(mid<query):
            high=position-1
    return -1
cards=[2,10,4,7,9]
cards.sort(reverse=True)
cards
```

```python
locate_card(cards,10)
```

Here my algorithm :

1. Uses binary search to find one occurrence.
2. Walk left one by one until the first occurrence is found.

Time complexity =O(logn)+O(n)

not optimized solution

The better approach:

Instead of walking left after finding the query, keep using binary search.

When you find the query:

- Save the position.
- Continue searching the left half to see if there's an earlier occurrence.

```python
def test_location(cards,query,position):
    tl=0   
    th=position
    while(tl<=th):
        mid=(tl+th)//2
        mid_num=cards[mid]
        print('low',tl,' mid_number',mid_num,' high',th,' mid',mid)
        if(mid_num==query):
            if(mid==0 or cards[mid-1]!=query):
                return mid
            th=mid-1
        elif(mid_num!=query):
            tl=mid+1
            
def locate_card(cards,query):   
    low=0
    high=len(cards)-1
    while low<=high:
        position=(low+high)//2
        mid_num=cards[position]
        print('low',low,' mid_number',mid_num,' high',high,' mid',position)
        if(mid_num==query):
            if(position>0 and cards[position-1]==query):
                position=test_location(cards,query,position)
            return position
        elif(mid_num>query):
            low=position+1
        elif(mid_num<query):
            high=position-1
    return -1

cards=[2,2,1,1,8,9,3,6,3,10,10,10, 9, 8]
cards.sort(reverse=True)
cards
```

```python
def test_cases(cards,mid,query):
    if(cards[mid]==query):
        if(mid-1>=0 and cards[mid-1]==query):
            return 'left'
        else:
            return 'found'
    elif(cards[mid]>query):
        return 'right'
    elif(cards[mid]<query):
        return 'left'


def locate_cards(cards,query):
    h=len(cards)-1
    l=0
    while(h>=l):
        mid=(h+l)//2
        print('low',l,'high',h,'mid',mid)
        result=test_cases(cards,mid,query)
        
        if(result=='found'):
            return mid
        elif(result == 'left'):
            h=mid-1
        elif(result=='right'):
            l=mid+1
    return -1
cards=[10,10,2,2,1,1,8,9,3,6,3,10,10,10, 9, 8]
cards.sort(reverse=True)
cards
```

```python
locate_cards(cards,10)
```
