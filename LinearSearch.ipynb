we can use brute force solution for checking all possible answer.

1. create a variable position=0.
2. check position in card equals to query
3. if it is equal then we get the ans so return it
4. if not equal increase position by 1 and repeat 2 and 3
5. if number not found till the end return -1

### linear search algorithm.

Implementation and testing using test cases.

```python
def locate_card(cards, query):
    position = 0
    for i in range(0, len(cards)):
        if(len(cards) == 0):
            return -1
        if(cards[position] == query):
            return position
        if(position == (len(cards)-1)):
            return -1
        if(cards[position] != query):
            position += 1
```
